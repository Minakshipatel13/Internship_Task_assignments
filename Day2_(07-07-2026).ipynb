##**Dictionary**

A dictionary is a collection of key-value pairs. It is mutable, unordered (Python 3.7+ preserves insertion order), and keys must be unique.

```python
# Dictionary Example

student = {
    "Name": "Minakshi",
    "Age": 20,
    "Branch": "Computer"
}

print("Original Dictionary:", student)

# 1. get()
print("Name:", student.get("Name"))

# 2. keys()
print("Keys:", student.keys())

# 3. values()
print("Values:", student.values())

# 4. update()
student.update({"City": "Surat"})
print("Update:", student)

# 5. pop()
student.pop("Age")
print("Pop:", student)
```

##**List**

A list is an ordered, mutable collection that allows duplicate values.

```python
# List Example

numbers = [10, 20, 30, 40]

print("Original List:", numbers)

# 1. append()
numbers.append(50)
print("append():", numbers)

# 2. extend()
numbers.extend([60, 70])
print("extend():", numbers)

# 3. insert()
numbers.insert(2, 25)
print("insert():", numbers)

# 4. remove()
numbers.remove(40)
print("remove():", numbers)

# 5. pop()
numbers.pop()
print("pop():", numbers)
```

##**Tuple**

A tuple is an ordered collection that is immutable (cannot be changed).

```python
# Tuple Example

t = (10, 20, 30, 20, 40)

print("Tuple:", t)

# 1. count()
print("Count of 20:", t.count(20))

# 2. index()
print("Index of 30:", t.index(30))

# 3. len()
print("Length:", len(t))

# 4. max()
print("Maximum:", max(t))

# 5. min()
print("Minimum:", min(t))
```

##**Set**

A set is an unordered collection of unique elements. It does not allow duplicates.

```python
# Set Example

s = {10, 20, 30}

print("Original Set:", s)

# 1. add()
s.add(40)
print("add():", s)

# 2. update()
s.update([50, 60])
print("update():", s)

# 3. remove()
s.remove(20)
print("remove():", s)

# 4. discard()
s.discard(100)
print("discard():", s)

# 5. pop()
removed = s.pop()
print("Removed:", removed)
print("pop():", s)
```

##**if** **Statement**

The if statement executes a block of code only when the condition is True.

```python
age = 20

if age >= 18:
    print("Eligible for Voting")
```

##**if-else**

Executes one block if the condition is True, otherwise another block.

```python
num = 5

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")
```

##**if-elif-else**

Used to check multiple conditions.

```python
marks = 75

if marks >= 90:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
else:
    print("Grade C")
```

##**Nested if-else**

An if statement inside another if statement.

```python
age = 22
citizen = True

if age >= 18:
    if citizen:
        print("Eligible to Vote")
    else:
        print("Not a Citizen")
else:
    print("Under Age")
```

##**Break**

The break statement stops the loop immediately.

```python
for i in range(1, 11):
    if i == 6:
        break
    print(i)
```

##**Continue**

The continue statement skips the current iteration.

```python
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

##**Pass**

The pass statement does nothing. It is used as a placeholder.

```python
for i in range(5):
    if i == 2:
        pass
    print(i)
```

##**Input**

The input() function takes input from the user.

```python
name = input("Enter Your Name: ")

print("Welcome", name)
```

##**range()**

The range() function generates a sequence of numbers.

```python
for i in range(1, 6):
    print(i)
```

##**len()**

The len() function returns the number of elements.

```python
fruits = ["Apple", "Banana", "Mango"]

print("Length:", len(fruits))
```

##**type()**

The type() function returns the data type of an object.

```python
a = 10
b = "Python"
c = [1, 2, 3]

print(type(a))
print(type(b))
print(type(c))
```

##**for Loop**

A for loop is used to iterate over a sequence.

```python
fruits = ["Apple", "Banana", "Mango"]

for fruit in fruits:
    print(fruit)
```

##**while Loop**

A while loop executes as long as the condition is True.

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```
