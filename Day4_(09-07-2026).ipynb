Observation:

The dataset contains a few missing values in the following columns:

  - Energy Type
  - Production (GWh)
  - Population

Since these missing values are very few compared to the total dataset size (2500 records), they can be handled without affecting the overall analysis.

Categorical values will be filled using the Mode, while numerical values will be filled using the Mean.

```python
# Handling Missing Values

df["Energy Type"] = df["Energy Type"].fillna(df["Energy Type"].mode()[0])

df["Production (GWh)"] = df["Production (GWh)"].fillna(df["Production (GWh)"].mean())

df["Population"] = df["Population"].fillna(df["Population"].mean())
```

  - The missing values have been successfully handled.

  - Mode was used for the categorical feature 'Energy Type' because it replaces missing values with the most frequent category.

  - Mean was used for the numerical features 'Production (GWh)' and 'Population' because it preserves the overall average of the dataset.

```python
print(df.isnull().sum())
```

  - All missing values have been successfully removed.

  - The dataset now contains no null values and is ready for generalization, feature relationship analysis, and further data exploration.
